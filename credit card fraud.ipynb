```python
#fraud cases : 1
#non-fraud cases : 0

fraud = data[data['Class'] == 1]
n_fraud = data[data['Class'] == 0]
print("fraud cases :", len(fraud))
print("non fraud cases :", len(n_fraud))
```

```python
data.isnull().sum()
```

```python
# Trying without pre-processing

x = data.drop("Class", axis = 1).values
y = data["Class"].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accRandF = accuracy_score(y_test, y_pred)*100
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

logR = LogisticRegression()
logR.fit(x_train, y_train)

y_pred = logR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accLogR = accuracy_score(y_test, y_pred)*100
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
from sklearn.linear_model import LinearRegression

linR = LinearRegression()
linR.fit(x_train, y_train)

y_pred = logR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accLinR = accuracy_score(y_test, y_pred)*100
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

tClass = DecisionTreeClassifier()
tClass.fit(x_train, y_train)

y_pred = tClass.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accDecT = accuracy_score(y_test, y_pred)*100
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
print("      Model          Accuracy(%)\n")
print("Linear Regression :  ", accLinR)
print("Logistic Regression :", accLogR)
print("Decision Tree :      ", accDecT)
print("Random Forest :      ", accRandF)
```
